```python
import pandas as pd

df = pd.read_csv("./salaries_by_college_major.csv")
# First 5 lines
df.head()
```

```python
# Shape of the data, (rows, columns)
df.shape
```

```python
# Names of all the columns
df.columns
```

```python
#Find all non number or NaN values
df.isna()
```

```python
# Print last 5 lines
df.tail()
```

```python
# Delete all Nan values
clean_df = df.dropna()
clean_df.tail()
```

```python
# Max value of a particular column
clean_df['Starting Median Salary'].max()
```

```python
#Index of that max value
clean_df['Starting Median Salary'].idxmax()
```

```python
# Locate the value of column, row (cell)
clean_df['Undergraduate Major'].loc[43]
```

```python
# Same as before
clean_df['Undergraduate Major'][43]
```

```python
# If no column specified, entire row given
clean_df.loc[43]
```

```python
# Print max values for particular columns, and then print the column that it is needed with that info
print(clean_df['Mid-Career Median Salary'].max())
clean_df['Undergraduate Major'].loc[clean_df['Starting Median Salary'].idxmax()]
```

```python
# Same buw with min
print(clean_df['Starting Median Salary'].min())
clean_df['Undergraduate Major'].loc[clean_df['Starting Median Salary'].idxmin()]
```

```python
# Print max values for particular columns, and then print the entire row
print(clean_df['Mid-Career Median Salary'].min())
clean_df.loc[clean_df['Mid-Career Median Salary'].idxmin()]
```

```python
# Arithmetic operations with columns
clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
# Insert in column 1, with title 'Spread', the entire column calculated before
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
# Sort df by the columns values
low_risk = clean_df.sort_values('Spread', ascending=False)
# See only this 2 columns in head output
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
# Sort df by the columns values
low_risk = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
# See only this 2 columns in head output
low_risk[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
# Count how many Groups we have in each column
clean_df.groupby('Group').count()
# Tell pandas to display only 2 digits in the float format
pd.options.display.float_format = '{:,.2f}'.format 
```

```python
# Mean salary by group
clean_df.groupby('Group').mean()
```
